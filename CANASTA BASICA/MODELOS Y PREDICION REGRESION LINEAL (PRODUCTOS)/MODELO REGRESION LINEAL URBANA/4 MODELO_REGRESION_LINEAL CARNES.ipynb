# __ETAPA 2: FILTRAR LOS DATOS DE LAS CARNES__

```python
# Filtrar solo el carnes
df_Carnes_1 = df[df["ARTICULO"] == "Carnes 1_ /"].copy()

#df.drop(267, inplace=True)
df_Carnes_1 = df_Carnes_1.drop(267, axis=0)

# Verificar el dataset filtrado

df_Carnes_1.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_Carnes_1' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_Carnes_1_melted = df_Carnes_1.melt(id_vars=["AÑO", "ARTICULO"],
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Carnes_1_melted["MES"] = df_Carnes_1_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_Carnes_1_melted["FECHA"] = pd.to_datetime(df_Carnes_1_melted["AÑO"].astype(str) + "-" + df_Carnes_1_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_Carnes_1_melted = df_Carnes_1_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Carnes_1_melted.reset_index(drop=True, inplace=True)

df_Carnes_1_melted.index = df_Carnes_1_melted.index + 1


# Mostrar los primeros registros
df_Carnes_1_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_Carnes_1_melted["FECHA_ORD"] = df_Carnes_1_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Carnes_1_melted[["FECHA_ORD"]]  # Matriz de características
y = df_Carnes_1_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python


# Convertir la columna FECHA a datetime (si no lo está)
df_Carnes_1_melted["FECHA"] = pd.to_datetime(df_Carnes_1_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Carnes_1_melted["FECHA_ORD"] = df_Carnes_1_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Carnes_1_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_Carnes_1_melted["PRECIO"].values       # Variable objetivo (precio de carnes)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_Carnes_1_melted["FECHA"], df_Carnes_1_melted["PRECIO"], color="blue", label="Precio Real", alpha=0.7)

# Línea de regresión en datos históricos
plt.plot(df_Carnes_1_melted["FECHA"], y_pred, color="red", linewidth=2, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="green", label="Predicciones 2025-2030", alpha=0.7)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_ticks = np.arange(0.2000, 0.6001, 0.0100)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio de las Carnes (2025-2030)")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__

Error Cuadrático Medio (MSE): 0.000324 = Este valor sigue siendo pequeño, lo cual indica que el modelo tiene un buen ajuste con los datos y el error es bajo.

Raíz del Error Cuadrático Medio (RMSE): 0.018008 = En promedio, el modelo tiene un error de aproximadamente 1.8 centavos en la predicción del precio. Es un error pequeño y muestra una buena precisión.

Error Absoluto Medio (MAE): 0.013766 = El modelo se equivoca, en promedio, por 1.38 centavos por predicción, lo que también refleja un buen desempeño, aunque ligeramente mayor que el RMSE.

Coeficiente de Determinación (R²): 0.950238 = Un R² de 95.02% indica que el modelo explica el 95.02% de la variabilidad en los precios, lo cual es excelente y muestra que el modelo tiene una alta capacidad para predecir los precios de las carnes con precisión.
