# __ETAPA 2: FILTRAR LOS DATOS DE GRASAS__

```python
# Filtrar solo el Pan Francés
df_Grasas_2 = df[df["ARTICULO"].str.strip() == "Grasas 2 _ /"].copy()

#df.drop(268, inplace=True)
df_Grasas_2 = df_Grasas_2.drop(268, axis=0)

# Verificar el dataset filtrado

df_Grasas_2.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_pan' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_Grasas_2_melted = df_Grasas_2.melt(id_vars=["AÑO", "ARTICULO"],
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Grasas_2_melted["MES"] = df_Grasas_2_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_Grasas_2_melted["FECHA"] = pd.to_datetime(df_Grasas_2_melted["AÑO"].astype(str) + "-" + df_Grasas_2_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_Grasas_2_melted = df_Grasas_2_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Grasas_2_melted.reset_index(drop=True, inplace=True)

df_Grasas_2_melted.index = df_Grasas_2_melted.index + 1


# Mostrar los primeros registros
df_Grasas_2_melted.head()
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Grasas_2_melted[["FECHA_ORD"]]  
y = df_Grasas_2_melted["PRECIO"]       
# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
from sklearn.linear_model import LinearRegression

# Convertir la columna FECHA a datetime (si no lo está)
df_Grasas_2_melted["FECHA"] = pd.to_datetime(df_Grasas_2_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Grasas_2_melted["FECHA_ORD"] = df_Grasas_2_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Grasas_2_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_Grasas_2_melted["PRECIO"].values       # Variable objetivo (precio de grasa)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_Grasas_2_melted["FECHA"], df_Grasas_2_melted["PRECIO"], color="b", label="Precio Real", alpha=0.7)

# Línea de regresión en datos históricos
plt.plot(df_Grasas_2_melted["FECHA"], y_pred, color="r", linewidth=2, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="g", label="Predicciones 2025-2030", alpha=0.7)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_ticks = np.arange(0.0200, 0.1501, 0.0050)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio de las Grasas (2025-2030)")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__
Error Cuadrático Medio (MSE): 0.000102 = Muy pequeño, lo cual es bueno.

Raíz del Error Cuadrático Medio (RMSE): 0.010093 = En promedio, el modelo tiene un error de aproximadamente 1.0 centavos en la predicción del precio.

Error Absoluto Medio (MAE): 0.008234 = Indica que, en promedio, el modelo se equivoca por 0.82 centavos por predicción.

Coeficiente de Determinación (R²): 0.710064 = Indica que el 71.0% de la variabilidad en el precio de la grasa se explica por la fecha.
