# __ETAPA 2: FILTRAR LOS DATOS DE LAS VERDURAS__

```python
df_verduras_4 = df[df["ARTICULO"] == "Verduras 4_ /"].copy()

# Eliminar la fila con índice 273 si existe
if 273 in df_verduras_4.index:
    df_verduras_4 = df_verduras_4.drop(273, axis=0)

# Verificar el dataset filtrado
df_verduras_4.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_verduras_4' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_verduras_4_melted = df_verduras_4.melt(id_vars=["AÑO", "ARTICULO"],
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_verduras_4_melted["MES"] = df_verduras_4_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_verduras_4_melted["FECHA"] = pd.to_datetime(df_verduras_4_melted["AÑO"].astype(str) + "-" + df_verduras_4_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_verduras_4_melted = df_verduras_4_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_verduras_4_melted.reset_index(drop=True, inplace=True)

df_verduras_4_melted.index = df_verduras_4_melted.index + 1


# Mostrar los primeros registros
df_verduras_4_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_verduras_4_melted["FECHA_ORD"] = df_verduras_4_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_verduras_4_melted[["FECHA_ORD"]]  # Matriz de características
y = df_verduras_4_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
from sklearn.linear_model import LinearRegression

# Convertir la columna FECHA a datetime (si no lo está)
df_verduras_4_melted["FECHA"] = pd.to_datetime(df_verduras_4_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_verduras_4_melted["FECHA_ORD"] = df_verduras_4_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_verduras_4_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_verduras_4_melted["PRECIO"].values       # Variable objetivo (precio del pan)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_verduras_4_melted["FECHA"], df_verduras_4_melted["PRECIO"], color="b", label="Precio Real", alpha=0.7)

# Línea de regresión en datos históricos
plt.plot(df_verduras_4_melted["FECHA"], y_pred, color="r", linewidth=3, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="g", label="Predicciones 2025-2030", alpha=0.2)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_ticks = np.arange(0.0000, 0.4001, 0.0200)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio de las Verduras (2025-2030)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

ANALISIS:

### **Resumen del Análisis**  

El modelo de regresión lineal muestra la tendencia histórica de los precios  y predice su evolución hasta 2030. La gráfica refleja fluctuaciones en los precios, con una tendencia general capturada por la línea de regresión. Las predicciones indican una posible continuidad de la tendencia observada, aunque pueden verse afectadas por factores externos.

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__

Error Cuadrático Medio (MSE): 0.000452 = Muy pequeño, lo que indica que el modelo tiene una baja cantidad de error cuadrático, lo cual es positivo.

Raíz del Error Cuadrático Medio (RMSE): 0.021253 = En promedio, el modelo tiene un error de aproximadamente 2.1 centavos en la predicción del precio de las verduras.

Error Absoluto Medio (MAE): 0.016027 = Indica que, en promedio, el modelo se equivoca por 1.6 centavos por predicción.

Coeficiente de Determinación (R²): 0.769639 = Indica que el 76.96% de la variabilidad en el precio de las verduras se explica por la fecha, lo que sugiere un buen ajuste del modelo.

## __Gráfico de Residuos__

Este gráfico nos ayuda a ver si el modelo tiene patrones en sus errores

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cálculo de los residuos
residuos = y - y_pred

# Crear gráfico de dispersión de residuos
plt.figure(figsize=(12,5))
plt.scatter(df_verduras_4_melted["FECHA"], residuos, color="b", alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--', linewidth=1.5)  # Línea de referencia

# Personalizar el gráfico
plt.xlabel("Fecha")
plt.ylabel("Error de Predicción (Residuo)")
plt.title("Análisis de Residuos")
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

ANALISIS:

El análisis de residuos indica la distribución de los errores del modelo de regresión lineal. Se observa que los residuos están dispersos en torno a la línea de referencia (cero), lo que sugiere que el modelo no presenta un sesgo sistemático. Sin embargo, si hay patrones visibles o acumulación de errores en ciertos períodos, esto podría indicar que la regresión lineal no captura completamente la variabilidad del precio de "verduras_4".

## __Análisis de Residuos__

    Los residuos (errores entre valores reales y predichos) están distribuidos en torno a la línea horizontal de referencia (0).

    Se observa un patrón no aleatorio en algunas secciones, lo cual sugiere que la regresión lineal no capta perfectamente todas las variaciones del precio de verduras.

    La dispersión de los residuos parece aumentar en años recientes, lo que indica que el modelo puede tener más dificultades para ajustarse a cambios recientes en la dinámica del precio.
