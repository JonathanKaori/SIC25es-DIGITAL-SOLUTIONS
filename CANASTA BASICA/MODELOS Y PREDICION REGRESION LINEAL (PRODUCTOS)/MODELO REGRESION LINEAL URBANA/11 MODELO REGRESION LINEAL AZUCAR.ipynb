```python
# Filtrar solo el Azucar
df_Azúcar = df[df["ARTICULO"] == "Azúcar"].copy()


#df.drop(264, inplace=True)
df_Azúcar = df_Azúcar.drop(274, axis=0)

# Verificar el dataset filtrado

df_Azúcar.head(25)

```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_Azúcar' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_Azúcar_melted = df_Azúcar.melt(id_vars=["AÑO", "ARTICULO"], 
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO", 
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Azúcar_melted["MES"] = df_Azúcar_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_Azúcar_melted["FECHA"] = pd.to_datetime(df_Azúcar_melted["AÑO"].astype(str) + "-" + df_Azúcar_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_Azúcar_melted = df_Azúcar_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Azúcar_melted.reset_index(drop=True, inplace=True)

df_Azúcar_melted.index = df_Azúcar_melted.index + 1


# Mostrar los primeros registros 
df_Azúcar_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_Azúcar_melted["FECHA_ORD"] = df_Azúcar_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Azúcar_melted[["FECHA_ORD"]]  # Matriz de características
y = df_Azúcar_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)


```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python
# Convertir la columna FECHA a datetime (si no lo está)
df_Azúcar_melted["FECHA"] = pd.to_datetime(df_Azúcar_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Azúcar_melted["FECHA_ORD"] = df_Azúcar_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Azúcar_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_Azúcar_melted["PRECIO"].values       # Variable objetivo (precio del pan)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_Azúcar_melted["FECHA"], df_Azúcar_melted["PRECIO"], color="blue", label="Precio Real", alpha=0.7)

# Línea de regresión en datos históricos
plt.plot(df_Azúcar_melted["FECHA"], y_pred, color="red", linewidth=2, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="green", label="Predicciones 2025-2030", alpha=0.7)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_ticks = np.arange(0.0400, 0.1001, 0.0050)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio del Azúcar (2025-2030)")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__

* Error Cuadrático Medio (MSE) = 0.000018 = Este valor es muy pequeño, lo que indica que las diferencias entre los valores predichos y los valores reales son mínimas en promedio.

* Raíz del Error Cuadrático Medio (RMSE) = 0.004206 = RMSE mide el error en las mismas unidades que los datos originales. Un valor de 0.0042 sugiere que las predicciones son bastante precisas, con un margen de error pequeño.

* Error Absoluto Medio (MAE) = 0.003416 = Representa el error promedio sin importar la dirección (positiva o negativa). Un MAE bajo indica que el modelo predice con una buena exactitud en la mayoría de los casos.

* Coeficiente de Determinación (R²) = 0.879533 = Un R² de 0.88 sugiere que el 87.95% de la variabilidad en los datos es explicada por el modelo. Esto indica un buen ajuste, aunque todavía hay un margen de mejora.

## __Gráfico de Residuos__

Este gráfico nos ayuda a ver si el modelo tiene patrones en sus errores

```python
import seaborn as sns

# Cálculo de los residuos
residuos = y - y_pred  

# Crear gráfico de dispersión de residuos
plt.figure(figsize=(12,5))
plt.scatter(df_Azúcar_melted["FECHA"], residuos, color="purple", alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1.5)  # Línea de referencia

# Personalizar el gráfico
plt.xlabel("Fecha")
plt.ylabel("Error de Predicción (Residuo)")
plt.title(" Análisis de Residuos")
plt.grid()
plt.show()
```

## __Análisis de Residuos__

 Definición de residuos: Son la diferencia entre los valores reales y los valores predichos por el modelo. Si el modelo es preciso, los residuos deberían distribuirse aleatoriamente alrededor de cero.

   ###  __Patrón observado:__

* Tendencias en los residuos: Se observan agrupaciones diagonales, lo que sugiere que los errores del modelo no son completamente aleatorios.

* Estructura repetitiva: Parece haber una relación temporal en los errores, lo que indica que el modelo podría no estar capturando del todo la dinámica del precio del azúcar.

* Mayor error en ciertas épocas: En algunos periodos (como 2016 y 2020) hay residuos más alejados de cero, lo que sugiere que el modelo tuvo más dificultades en esas fechas.

## __Histograma de Errores__ 

Este gráfico muestra cómo están distribuidos los errores del modelo.

```python
# Crear histograma de errores
plt.figure(figsize=(10,5))
sns.histplot(residuos, bins=20, kde=True, color="orange")

# Personalizar
plt.xlabel("Error de Predicción (Residuo)")
plt.ylabel("Frecuencia")
plt.title("Distribución de Errores del Modelo")
plt.grid()
plt.show()
```

## __Distribución de Errores del Modelo__

* Forma aproximadamente simétrica: Los residuos están centrados cerca de 0, lo que sugiere que el modelo no tiene un sesgo sistemático (ni sobreestima ni subestima significativamente).

* Distribución ligeramente sesgada a la derecha: Hay más errores positivos que negativos, lo que indica que el modelo tiende a predecir valores ligeramente más bajos de lo real.

* Mayor concentración en torno a 0: La mayoría de los errores son pequeños, lo cual es una buena señal de precisión.
