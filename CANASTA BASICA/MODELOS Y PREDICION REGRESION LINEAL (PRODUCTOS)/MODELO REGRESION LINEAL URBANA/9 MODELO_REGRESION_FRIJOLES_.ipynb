# __ETAPA 2: FILTRAR LOS DATOS DE FRIJOLES__

```python
# Filtrar solo el Pan Francés
df_Frijoles = df[df["ARTICULO"] == "Frijoles"].copy()


#df.drop(264, inplace=True)
df_Frijoles = df_Frijoles.drop(272, axis=0)

# Verificar el dataset filtrado

df_Frijoles.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_Frijoles' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_Frijoles_melted = df_Frijoles.melt(id_vars=["AÑO", "ARTICULO"],
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Frijoles_melted["MES"] = df_Frijoles_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_Frijoles_melted["FECHA"] = pd.to_datetime(df_Frijoles_melted["AÑO"].astype(str) + "-" + df_Frijoles_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_Frijoles_melted = df_Frijoles_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Frijoles_melted.reset_index(drop=True, inplace=True)

df_Frijoles_melted.index = df_Frijoles_melted.index + 1


# Mostrar los primeros registros
df_Frijoles_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_Frijoles_melted["FECHA_ORD"] = df_Frijoles_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Frijoles_melted[["FECHA_ORD"]]  # Matriz de características
y = df_Frijoles_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python


# Convertir la columna FECHA a datetime (si no lo está)
df_Frijoles_melted["FECHA"] = pd.to_datetime(df_Frijoles_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Frijoles_melted["FECHA_ORD"] = df_Frijoles_melted["FECHA"].map(lambda x: x.toordinal())


# Definir variables para regresión
X = df_Frijoles_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_Frijoles_melted["PRECIO"].values       # Variable objetivo (precio del pan)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))


# Línea que une los datos históricos (menos opaca)
plt.plot(df_Frijoles_melted["FECHA"], df_Frijoles_melted["PRECIO"], 
         color="blue", linestyle="--", alpha=0.5, linewidth=1.5)

# Datos históricos (puntos más grandes con borde negro)
plt.scatter(df_Frijoles_melted["FECHA"], df_Frijoles_melted["PRECIO"], 
            color="blue", edgecolors="b", s=50, label="Precio Real", alpha=0.9)




# Línea de regresión en datos históricos
plt.plot(df_Frijoles_melted["FECHA"], y_pred, color="red", linewidth=3, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="green", label="Predicciones 2025-2030", alpha=0.7)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_ticks = np.arange(0.0500, 0.2501, 0.0100)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio de Frijoles (2025-2030)")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__

*Error Cuadrático Medio (MSE): 0.000982 = Muy pequeño, lo cual es bueno.

Raíz del Error Cuadrático Medio (RMSE): 0.031330 = En promedio, el modelo tiene un error de aproximadamente 3.1 centavos en la predicción del precio de.

Error Absoluto Medio (MAE): 0.024235 = Indica que, en promedio, el modelo se equivoca por 2.4 centavos por predicción.

Coeficiente de Determinación (R²): 0.557361 = Indica que el 55.7% de la variabilidad en el precio de Frijoles se explica por la fecha.

## __Gráfico de Residuos__

Este gráfico nos ayuda a ver si el modelo tiene patrones en sus errores

```python
import seaborn as sns

# Cálculo de los residuos
residuos = y - y_pred

# Crear gráfico de dispersión de residuos
plt.figure(figsize=(12,5))
plt.scatter(df_Frijoles_melted["FECHA"], residuos, color="b", alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2.5)  # Línea de referencia

# Personalizar el gráfico
plt.xlabel("Fecha")
plt.ylabel("Error de Predicción (Residuo)")
plt.title(" Análisis de Residuos")
plt.grid()
plt.show()
```

## __Análisis de Residuos__

    Los residuos (errores entre valores reales y predichos) están distribuidos en torno a la línea horizontal de referencia (0).

    Se observa un patrón no aleatorio en algunas secciones, lo cual sugiere que la regresión lineal no capta perfectamente todas las variaciones del precio de los frijoles.

    La dispersión de los residuos parece aumentar en años recientes, lo que indica que el modelo puede tener más dificultades para ajustarse a cambios recientes en la dinámica del precio.

## __Histograma de Errores__

## __Distribución de Errores del Modelo__

    Sin embargo, se observan algunos valores extremos en ambos lados, lo que indica que en ciertos casos el modelo comete errores más grandes.
