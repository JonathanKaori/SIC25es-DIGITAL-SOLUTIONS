# __ETAPA 2: FILTRAR LOS DATOS DE LECHE FLUIDA__

```python
# Filtrar solo el LECHE
df_Leche_Fluida = df[df["ARTICULO"] == "Leche Fluida *"].copy()


#df.drop(270, inplace=True)
df_Leche_Fluida = df_Leche_Fluida.drop(270, axis=0)

# Verificar el dataset filtrado

df_Leche_Fluida.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_Leche_Fluida' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_Leche_Fluida_melted = df_Leche_Fluida.melt(id_vars=["AÑO", "ARTICULO"],
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Leche_Fluida_melted["MES"] = df_Leche_Fluida_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_Leche_Fluida_melted["FECHA"] = pd.to_datetime(df_Leche_Fluida_melted["AÑO"].astype(str) + "-" + df_Leche_Fluida_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_Leche_Fluida_melted = df_Leche_Fluida_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Leche_Fluida_melted.reset_index(drop=True, inplace=True)

df_Leche_Fluida_melted.index = df_Leche_Fluida_melted.index + 1


# Mostrar los primeros registros
df_Leche_Fluida_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_Leche_Fluida_melted["FECHA_ORD"] = df_Leche_Fluida_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Leche_Fluida_melted[["FECHA_ORD"]]  # Matriz de características
y = df_Leche_Fluida_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
from sklearn.linear_model import LinearRegression

# Convertir la columna FECHA a datetime (si no lo está)
df_Leche_Fluida_melted["FECHA"] = pd.to_datetime(df_Leche_Fluida_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Leche_Fluida_melted["FECHA_ORD"] = df_Leche_Fluida_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Leche_Fluida_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_Leche_Fluida_melted["PRECIO"].values       # Variable objetivo (precio de la leche)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_Leche_Fluida_melted["FECHA"], df_Leche_Fluida_melted["PRECIO"], color="b", label="Precio Real", alpha=0.7)

# Línea de regresión en datos históricos
plt.plot(df_Leche_Fluida_melted["FECHA"], y_pred, color="r", linewidth=2, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="g", label="Predicciones 2025-2030", alpha=0.7)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_ticks = np.arange(0.0500, 0.1501, 0.0050)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio de la Leche Fluida (2025-2030)", fontsize=16)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)  # Cuadrícula con líneas discontinuas y opacidad
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

ANALISIS:

Este gráfico proporciona una visualización clara tanto de los datos históricos como de las predicciones futuras de los precios de la leche fluida. La regresión lineal utilizada ofrece una aproximación al comportamiento esperado de los precios, lo cual puede ser útil para la toma de decisiones estratégicas en relación con la planificación y fijación de precios en el sector de la leche.

Las predicciones para 2025-2031 pueden ser ajustadas en función de nuevas variaciones o cambios en el mercado, y su precisión dependerá de cómo se mantenga el comportamiento histórico del precio de la leche.

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__

Error Cuadrático Medio (MSE): 0.000047 = Muy pequeño, lo cual es positivo. Un valor cercano a cero indica que el modelo realiza predicciones con un error reducido.

Raíz del Error Cuadrático Medio (RMSE): 0.006823 = En promedio, el modelo tiene un error de aproximadamente 0.68 centavos por predicción. Esto significa que las predicciones están bastante cerca de los valores reales.

Error Absoluto Medio (MAE): 0.005409 = Indica que, en promedio, el modelo se equivoca por 0.54 centavos por predicción, lo que refleja un rendimiento bastante bueno.

Coeficiente de Determinación (R²): 0.647418 = Esto significa que aproximadamente el 64.7% de la variabilidad en el precio de la leche se explica por la fecha. Aunque el modelo explica una buena parte de la variabilidad, aún hay un 35.3% de variación que no se explica, lo que podría mejorar con ajustes en el modelo o con más datos.

 El gráfico de residuos muestra el análisis de los errores de predicción de un modelo de regresión lineal sobre el precio de la leche fluida.
