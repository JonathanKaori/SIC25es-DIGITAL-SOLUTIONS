# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Filtrar solo los datos de las Tortillas
df_Tortillas = df[df["ARTICULO"] == "Tortillas"].copy()

# Transformar de formato ancho (columnas de meses) a formato largo (fecha única)
df_Tortillas_melted = df_Tortillas.melt(id_vars=["AÑO", "ARTICULO"],
                                        value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
                                                    "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                                        var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Tortillas_melted["MES"] = df_Tortillas_melted["MES"].map(meses_dict)

# Crear la columna de fecha en formato datetime
df_Tortillas_melted["FECHA"] = pd.to_datetime(df_Tortillas_melted["AÑO"].astype(str) + "-" + df_Tortillas_melted["MES"].astype(str) + "-01")

# Filtrar solo los datos hasta el año 2024
df_Tortillas_melted = df_Tortillas_melted[df_Tortillas_melted["FECHA"] < "2025-01-01"]

# Ordenar por fecha y eliminar columnas innecesarias
df_Tortillas_melted = df_Tortillas_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Tortillas_melted.reset_index(drop=True, inplace=True)

# Mostrar los primeros registros
df_Tortillas_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_Tortillas_melted["FECHA_ORD"] = df_Tortillas_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Tortillas_melted[["FECHA_ORD"]]  # Matriz de características
y = df_Tortillas_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python


# Convertir la columna FECHA a datetime (si no lo está)
df_Tortillas_melted["FECHA"] = pd.to_datetime(df_Tortillas_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Tortillas_melted["FECHA_ORD"] = df_Tortillas_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Tortillas_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_Tortillas_melted["PRECIO"].values       # Variable objetivo (precio de Tortillas)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_Tortillas_melted["FECHA"], df_Tortillas_melted["PRECIO"], color="blue", label="Precio Real", alpha=0.7)

# Línea de regresión en datos históricos
plt.plot(df_Tortillas_melted["FECHA"], y_pred, color="red", linewidth=2, label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos, color="green", label="Predicciones 2025-2030", alpha=0.7)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_ticks = np.arange(0.1000, 0.3001, 0.0100)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha")
plt.ylabel("Precio ($)")
plt.title("Evolución y Predicción del Precio de Tortillas (2025-2030)")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 ANALISIS:

 Este análisis muestra cómo un modelo de regresión lineal puede ser utilizado para predecir los precios futuros de las tortillas basándose en los datos históricos que se logran optener.

  Aunque el modelo proporciona una proyección de los precios para el periodo de 2025-2031, es importante considerar que la regresión lineal asume que las tendencias pasadas seguirán de manera constante, lo cual podría no reflejar cambios bruscos o eventos y aumentos inesperados.
  
   El gráfico ayuda a visualizar tanto los datos reales como las predicciones futuras, facilitando el análisis de tendencias y posibles fluctuaciones.

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__

* Error Cuadrático Medio (MSE): 0.000257 = Este valor es relativamente pequeño, lo que indica que el modelo tiene un buen desempeño al minimizar el error cuadrático.

Raíz del Error Cuadrático Medio (RMSE): 0.016038 = En promedio, el modelo tiene un error de aproximadamente 1.6 centavos en la predicción del precio de tortillas.

Error Absoluto Medio (MAE): 0.012238 = Indica que, en promedio, el modelo se equivoca por 1.22 centavos por predicción.

Coeficiente de Determinación (R²): 0.854772 = Esto sugiere que el 85.5% de la variabilidad en el precio del pan se explica por la fecha, lo cual indica un buen ajuste del modelo a los datos.

## __Gráfico de Residuos__

Este gráfico nos ayuda a ver si el modelo tiene patrones en sus errores

```python
import seaborn as sns

# Cálculo de los residuos
residuos = y - y_pred

# Crear gráfico de dispersión de residuos
plt.figure(figsize=(12,5))
plt.scatter(df_Tortillas_melted["FECHA"], residuos, color="purple", alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1.5)  # Línea de referencia

# Personalizar el gráfico
plt.xlabel("Fecha")
plt.ylabel("Error de Predicción (Residuo)")
plt.title(" Análisis de Residuos")
plt.grid()
plt.show()
```

## __Análisis de Residuos__

  Este análisis de residuos permite evaluar la calidad del modelo de regresión. Si los residuos se distribuyen aleatoriamente alrededor de la línea de referencia, el modelo es adecuado. Sin embargo, patrones o residuos grandes pueden indicar problemas con el ajuste, como la necesidad de un modelo diferente o la presencia de valores atípicos.

## __Histograma de Errores__

Este gráfico muestra cómo están distribuidos los errores del modelo.

```python
# Crear histograma de errores
plt.figure(figsize=(10,5))
sns.histplot(residuos, bins=20, kde=True, color="orange")

# Personalizar
plt.xlabel("Error de Predicción (Residuo)")
plt.ylabel("Frecuencia")
plt.title("Distribución de Errores del Modelo")
plt.grid()
plt.show()
```

El histograma y la curva KDE proporcionan una visión clara de la distribución de los errores del modelo. Una distribución normal centrada en cero indica que el modelo tiene un buen ajuste. Sin embargo, si la distribución es sesgada o tiene outliers, podría ser necesario ajustar el modelo, considerar transformaciones de las variables o probar técnicas más complejas.

## __Distribución de Errores del Modelo__

El análisis de la distribución de errores del modelo es fundamental para verificar que el modelo esté realizando predicciones correctas y no esté sesgado o influenciado por patrones no capturados. Si los errores son aleatorios y centrados en cero, eso indica que el modelo es adecuado. Si hay patrones o sesgos, es posible que el modelo necesite ajustes o transformaciones adicionales.
