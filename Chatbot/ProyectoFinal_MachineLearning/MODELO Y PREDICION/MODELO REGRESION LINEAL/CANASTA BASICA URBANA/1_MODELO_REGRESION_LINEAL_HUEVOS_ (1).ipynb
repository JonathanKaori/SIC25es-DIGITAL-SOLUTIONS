# __ETAPA 2: FILTRAR LOS DATOS DEL PAN FRANCES__

```python
# Filtrar solo Huevos
df_Huevos = df[df["ARTICULO"] == "Huevos"].copy()


#df.drop(269, inplace=True)
df_Huevos = df_Huevos.drop(269, axis=0)

# Verificar el dataset filtrado

df_Huevos.head(25)
```

# __ETAPA 3: TRANFORMAR LOS DATOS EN UN FORMATO ANCHO A LARGO__

```python
import pandas as pd

# Asegúrate de cargar el DataFrame 'df_Huevos' antes de ejecutar este código

# Transformar de formato ancho (varias columnas de meses) a largo (fecha única)
df_Huevos_melted = df_Huevos.melt(id_vars=["AÑO", "ARTICULO"],
                            value_vars=["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
                                        "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"],
                            var_name="MES", value_name="PRECIO")

# Mapeo de nombres de meses a números
meses_dict = {
    "ENERO": 1, "FEBRERO": 2, "MARZO": 3, "ABRIL": 4, "MAYO": 5, "JUNIO": 6,
    "JULIO": 7, "AGOSTO": 8, "SEPTIEMBRE": 9, "OCTUBRE": 10, "NOVIEMBRE": 11, "DICIEMBRE": 12
}

# Convertir los nombres de los meses a números
df_Huevos_melted["MES"] = df_Huevos_melted["MES"].map(meses_dict)

# Crear la columna de fecha
df_Huevos_melted["FECHA"] = pd.to_datetime(df_Huevos_melted["AÑO"].astype(str) + "-" + df_Huevos_melted["MES"].astype(str) + "-01")

# Ordenar por fecha y eliminar columnas innecesarias
df_Huevos_melted = df_Huevos_melted.sort_values("FECHA")[["FECHA", "PRECIO"]]

# Reiniciar el índice
df_Huevos_melted.reset_index(drop=True, inplace=True)

df_Huevos_melted.index = df_Huevos_melted.index + 1


# Mostrar los primeros registros
df_Huevos_melted.head()
```

# __ETAPA 5: DIVISION DEL CONJUNTO DE DATOS__

```python
df_Huevos_melted["FECHA_ORD"] = df_Huevos_melted["FECHA"].map(lambda x: x.toordinal())
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y dependientes (y)
X = df_Huevos_melted[["FECHA_ORD"]]  # Matriz de características
y = df_Huevos_melted["PRECIO"]       # Variable objetivo

# División en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# __ETAPA 6: ENTRENAMIENTO DEL MODELO DE REGRESION LINEAL__

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)

print(f"Pendiente (m): {modelo.coef_[0]}")
print(f"Intercepto (b): {modelo.intercept_}")

y_pred = modelo.predict(X_test)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
from sklearn.linear_model import LinearRegression

# Convertir la columna FECHA a datetime
df_Huevos_melted["FECHA"] = pd.to_datetime(df_Huevos_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Huevos_melted["FECHA_ORD"] = df_Huevos_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Huevos_melted[["FECHA_ORD"]].values  # Característica (fecha en ordinal)
y = df_Huevos_melted["PRECIO"].values       # Variable objetivo (precio de huevos)

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Graficar datos reales y regresión lineal
plt.figure(figsize=(12,6))

# Datos históricos (color azul claro)
plt.scatter(df_Huevos_melted["FECHA"], df_Huevos_melted["PRECIO"],
            color="dodgerblue", label="Precio Real", alpha=0.7, s=50)

# Línea de regresión (color naranja)
plt.plot(df_Huevos_melted["FECHA"], y_pred, color="darkorange", linewidth=2, label="Regresión Lineal")

# Predicciones futuras (color verde oscuro)
plt.scatter(fechas_futuras, precios_predichos,
            color="seagreen", label="Predicciones 2025-2030", alpha=0.7, s=60)

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())  # Mostrar cada año
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))  # Formato Año

# Definir el rango del eje Y
y_ticks = np.arange(0.0600, 0.1901, 0.0050)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha", fontsize=12)
plt.ylabel("Precio ($)", fontsize=12)
plt.title("Evolución y Predicción del Precio de Huevos (2025-2030)", fontsize=14, fontweight="bold")

# Mejorar visualización
plt.legend()
plt.xticks(rotation=45)
plt.grid(alpha=0.4)  # Grid más suave
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
from sklearn.linear_model import LinearRegression

# Convertir la columna FECHA a datetime
df_Huevos_melted["FECHA"] = pd.to_datetime(df_Huevos_melted["FECHA"])

# Convertir la FECHA a formato ordinal para regresión
df_Huevos_melted["FECHA_ORD"] = df_Huevos_melted["FECHA"].map(lambda x: x.toordinal())

# Definir variables para regresión
X = df_Huevos_melted[["FECHA_ORD"]].values
y = df_Huevos_melted["PRECIO"].values

# Entrenar modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Generar predicciones sobre los datos históricos
y_pred = modelo.predict(X)

# Predecir precios futuros (2025-2030)
fechas_futuras = pd.date_range(start="2025-01-01", end="2030-12-01", freq="ME")
fechas_futuras_ordinal = np.array([fecha.toordinal() for fecha in fechas_futuras]).reshape(-1, 1)
precios_predichos = modelo.predict(fechas_futuras_ordinal)

# Establecer un estilo seguro
plt.style.use("ggplot")  # Alternativa segura

# Graficar
plt.figure(figsize=(12,6))

# Datos históricos
plt.scatter(df_Huevos_melted["FECHA"], df_Huevos_melted["PRECIO"],
            color="darkorange", label="Precio Real", alpha=0.8, s=60, edgecolors="black")

# Línea de regresión
plt.plot(df_Huevos_melted["FECHA"], y_pred,
         color="dodgerblue", linewidth=2.5, linestyle="-", label="Regresión Lineal")

# Predicciones futuras
plt.scatter(fechas_futuras, precios_predichos,
            color="mediumorchid", label="Predicciones 2025-2030", alpha=0.8, s=70, edgecolors="black")

# Personalizar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

# Rango del eje Y
y_ticks = np.arange(0.0600, 0.1901, 0.0050)
plt.yticks(y_ticks)

# Personalizar la gráfica
plt.xlabel("Fecha", fontsize=13)
plt.ylabel("Precio ($)", fontsize=13)
plt.title("Evolución y Predicción del Precio de Huevos (2025-2030)", fontsize=15, fontweight="bold", color="black")

plt.legend(frameon=True, shadow=True, fontsize=11)
plt.xticks(rotation=45)
plt.grid(alpha=0.5, linestyle="--")
plt.tight_layout()
plt.show()
```

# __ETAPA 7: EVALUACION DEL MODELO__

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calcular métricas de evaluación
mse = mean_squared_error(y, y_pred)  # Error cuadrático medio
rmse = np.sqrt(mse)  # Raíz del error cuadrático medio
mae = mean_absolute_error(y, y_pred)  # Error absoluto medio
r2 = r2_score(y, y_pred)  # Coeficiente de determinación

# Imprimir resultados
print(f" Error Cuadrático Medio (MSE): {mse:.6f}")
print(f" Raíz del Error Cuadrático Medio (RMSE): {rmse:.6f}")
print(f" Error Absoluto Medio (MAE): {mae:.6f}")
print(f" Coeficiente de Determinación (R²): {r2:.6f}")
```

## __Análisis de la Evaluación del Modelo__
Error Cuadrático Medio (MSE): 0.000040 = Este valor muy pequeño indica que el modelo tiene un error cuadrático mínimo, lo cual es excelente para la precisión de las predicciones.

Raíz del Error Cuadrático Medio (RMSE): 0.006313 = En promedio, el modelo tiene un error de aproximadamente 0.63 centavos por predicción del precio de los huevos, lo que refleja una alta precisión.

Error Absoluto Medio (MAE): 0.005177 = El modelo se equivoca, en promedio, por solo 0.52 centavos por predicción, lo que demuestra un buen desempeño en la predicción de precios.

Coeficiente de Determinación (R²): 0.829940 = Esto significa que aproximadamente el 82.99% de la variabilidad en el precio de los huevos se puede explicar mediante el modelo de regresión basado en las fechas.

Los residuos (errores de predicción) están distribuidos alrededor de la línea de referencia (y=0), lo que indica que el modelo tiene una buena capacidad de predicción, sin sesgos sistemáticos a lo largo del tiempo. Aunque la mayoría de los residuos son pequeños, algunos puntos se desvían significativamente, lo que podría indicar outliers o factores no capturados por el modelo.
