# __OBTENCION DE DATOS__

## __La base de datos de la canasta basica rural y urbana abarca un periodo desde el 2001 hasta enero del 2025__

## la obtencion de los datos fue a travez de tecnicas de web scraping y la consulta de diversos sitios web ya que la descarga directa de los sitios oficiales no era directamente en algun formato .csv , ni formato .xls y no se trataban tampoco de algun archivo Excel nativo, sino un archivo en formato HTML con tablas que Excel puede abrir, pero no es un verdadero archivo de Excel. Esto explicaba por qué al querer abrirlo en Jupyter Notebook se mostraba código HTML en lugar de datos tabulares.

### Los datos de "La Canasta Basica Alimentaria de El Salvador rural" fueron obtenidos del sitio web de la OFICINA NACIONAL DEL ESTADISTICA Y CENSO DE EL SALVADOR (ONEC-BCR)

### *https://onec.bcr.gob.sv/ipc_consulta/ipc_dat.asp*

```python
from bs4 import BeautifulSoup
import pandas as pd
import io  # Importar StringIO

# Cargar el archivo como texto
with open("descarga (1).xls", "r", encoding="utf-8") as f:
    html_content = f.read()

# Parsear el contenido HTML con BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Extraer todas las tablas
tables = soup.find_all("table")

# Convertir la primera tabla en un DataFrame 
df = pd.read_html(io.StringIO(str(tables[0])))[0]

# Mostrar los primeros registros
df.head()

# Guardar en CSV 
#df.to_csv("canasta_basica_rural .csv", index=False)
```
